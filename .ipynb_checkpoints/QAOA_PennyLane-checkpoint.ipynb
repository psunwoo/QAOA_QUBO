# Functions to produce Hamiltonians

$\bullet \textbf{Q_to_ Ising}$: <br>
converts the Q matrix in $\sum_i Q_{ii}x_i + \sum_{i,j}Q_{ij}x_ix_j$ with $x \in \{0,1\}$ into Ising Hamiltonian $H = \sum_ih_iZ_i +\sum_{i,j}J_{i,j}Z_iZ_j$ with $z \in\{1,-1\}$ via $\displaystyle x_i = \frac{1-Z_i}{2}$. Then in terms of $\{Z_i\}$, the Hamiltonian to find the ground state would look like: $Q_{ii}\left(\frac{1-Z_i}{2} \right) + \frac{Q_{ij}}{4}(1-Z_i - Z_j + Z_iZ_j)$

To give: $\displaystyle h_i = -\frac{Q_{ii}}{2} - \sum_j\frac{Q_{ij}}{4}$ and $\displaystyle J_{ij} = \frac{Q_{ij}}{4} $ with the c number energy offset of $\displaystyle \sum_i \frac{Q_{ii}}{2} + \sum_{i,j}\frac{Q_{ij}}{4}$.
<br><Br>
$\bullet \textbf{create_Hamiltonian}$:<br>
Creates the Ising Hamiltonian from the parameters.

```python
def Q_to_Ising(Q, offset = 0):
  """"Convert the upper triangular matrix in QUBO into element in J and h. Also constructs the operator lists in the right order as coefficients for later convenience"""
  num_qubits = len(Q)
  h = defaultdict(int)
  J = defaultdict(int)
  oneq_operators = []
  twoq_operators = []

  # Outer loop to sum over one index - 'i'
  for i in range(num_qubits):
    h[(i,)] -= Q[i,i]/2
    offset += Q[i,i]/2
    oneq_operators.append(qml.PauliZ(i))

  # Inner loop to sum over the other index - 'j'
    for j in range(i+1, num_qubits):
      J[(i,j)] = Q[i,j]/4
      h[(i,)] -= Q[i,j]/4
      h[(j,)] -= Q[i,j]/4
      offset += Q[i,j]/4
      twoq_operators.append(qml.PauliZ(i) @ qml.PauliZ(j))
  return h,J,offset, oneq_operators, twoq_operators

def create_cost_Hamiltonian(h=None, J=None, offset=0, oneq_operators = None, twoq_operators = None, Q = None, converted = 0):
  """ Create the Ising Hamiltonian from the parameters"""

  # Unless desired to be otherwise, carry out 'Q_to_Ising' within this function
  if converted == 0:
    if Q is None:
      raise ValueError("Q must be provided if converted = 0")

    h,J,offset, oneq_operators, twoq_operators = Q_to_Ising(Q, offset = offset)

  # If desired to carry out 'Q_to_Ising' separately
  if converted == 1:
    if (oneq_operators is None) or (twoq_operators is None):
      raise ValueError(f"No valid operator list form for QUBO is given. Please carry out {Q_to_Ising.__name__} first ")

  # Create the Ising Hamiltonian
  coeffs = list(h.values()) + list(J.values())
  coeff_max = max(abs(c) for c in coeffs) # This will be used for QAOA layer construction later
  operators = oneq_operators + twoq_operators
  return qml.Hamiltonian(coeffs, operators), offset, coeff_max

def create_mixer_Hamiltonian(num_qubits):
  coefficients = -np.ones(num_qubits)
  operators = []
  for i in range(num_qubits):
    operators.append(qml.PauliX(i))
  return qml.Hamiltonian(coefficients, operators)
```

# QAOA - Adiabatic (without the parameter optimization)

QAOA to mimic the adiabatic time evolution in a Trotterized fashion.  The Hamiltonian constructed above gives the cost function $H_C$, which at each Trotterized state, will give the time evolution $U(H_C, \gamma_k) = e^{-i\gamma_kH_C}$.  From the commuting nature of $Z$ operators, this evolution can be achieved by successive application of $RZZ(2\gamma_iJ_{ij})$ and $RZ(2\gamma_ih_i)$. On the other hand, the mixer operator leads to $U(X, \beta_k) = e^{i\beta_kX}$ with $X = H_M = \sum_j\sigma_{x,j}$, which is $RX(-2\beta_k)$.

Start from $\beta_0 = 1$ and $\gamma_0 = 0$. Then gradually decrease/increase $\beta / \gamma$, such that the ground state of the cost function is reached adibatically at the end of the algorithm.

$\textbf{Note that the time evolution in each Trotterized block happens for } t = \hbar$ <br>
$\textbf{Also note that, although with the number of steps converging to infinity, the adiabatic QAOA must converge to QA, in practice this does not happen due to rounding errors from each step.}$

```python
wires = len(Q)
device = qml.device('default.qubit', wires = range(wires))
# H_C = qml.Hamiltonian((1 / max(abs(np.linalg.eigvals(Q)))) * np.array(cost_Hamiltonian.coeffs), cost_Hamiltonian.ops)
H_M = create_mixer_Hamiltonian(wires)
depth = 5

def qaoa_layer(gamma, beta):
    qaoa.cost_layer(gamma, H_C)
    qaoa.mixer_layer(beta, H_M)

# Define circuit explicitly over depth
def circuit(params, **kwargs):
    """QAOA circuit with multiple layers applied."""
    for w in range(wires):
        qml.Hadamard(w)  # Initialize the state
    for d in range(depth):  # Explicit loop over layers
        qaoa_layer(params[0][d], params[1][d])

@qml.qnode(device)
def cost_function(params, normalized = True):
    circuit(params)
    if normalized == True:
      expval = qml.expval(H_C)  # Ensure `H_C` is differentiable
    else:
      expval = qml.expval(cost_Hamiltonian)
    return expval

@qml.qnode(device)
def prob_populations(params):
  circuit(params)
  return qml.probs(wires = range(wires))

def calculate_cost(spin_config):
  spin_config = np.array(list(spin_config), dtype=int)
  return spin_config.T @ Q @ spin_config
```

```python
# gammas and betas are produced with the given ansatz below to achieve accurate and fast convergence for the adiabatic evolution.
gammas = np.linspace(0,1,depth) # Start slow, accelerate
betas = np.linspace(0,0.8,depth, endpoint = False)[::-1]  # Start fast, slow down
parameters = [gammas, betas]
```

```python
probs = prob_populations(parameters)
plt.style.use('seaborn-v0_8')
plt.bar(range(2**wires), probs)
plt.xlabel("Qubit Configuration")
plt.ylabel("Probability")
plt.ylim(0, max(probs)*1.1)
plt.show()
```

```python
# Find the number of non-zero elements in probs
count = sum([1 for x in probs if x != 0])
print(count)

# Get top (N) indices sorted by highest probability
indices = np.where(probs >= nth_largest(probs, count))[0]
indices = indices[np.argsort(probs[indices])[::-1]]  # Sort by descending probabilities

# Convert to binary representation
binary_indices = [np.binary_repr(i, width=wires) for i in indices]

# Compute costs efficiently using list comprehension
costs = np.array([calculate_cost(b) for b in binary_indices])

# Get the 3 lowest costs and corresponding configurations
threshold = -nth_largest(-costs, 3)
final_mask = costs <= threshold

final_indices = np.array(binary_indices)[final_mask]
final_costs = costs[final_mask]
final_probs = probs[indices[final_mask].astype(int)]  # Get corresponding probabilities

# Print the results
for config, cost, prob in zip(final_indices, final_costs, final_probs):
    print(f"config: {config}, cost: {cost}, prob: {100*prob:.8f}%")

# Find the expectation value of the cost function from the algorithm
cost = cost_function(parameters)
print(cost)
```

```python
# Sum all the probabilities to yield the correct value (without considering the slack variables)
def check_first_five_bits(num, total_bits = 11, ref = 0b11110):
  first_five = (num >> (total_bits - 5))
  return first_five == ref

total_prob = 0
for index, prob in enumerate(probs):
  if check_first_five_bits(index):
    total_prob += prob
print(f"The probability to get the correct value is: {100 *total_prob}%")
```

# QAOA - Diabatic (with the parameter optimization)

Rather than mimicing adiabatic Quantum Annealing from the Trotterized gates, let us now employ the classical optimizer that runs on CPU/GPU (preferably).

```python
depth = 10 # Reduced depth for faster optimization in the diabatic case.

# Classical Optimization setup
steps = 200 # Number of optimization steps
optimizer = qml.AdamOptimizer(stepsize = 0.2)

# Randomly initialize parameters
params = np.array(np.stack([np.linspace(0, np.pi/2, depth), np.linspace(0, np.pi/4, depth, endpoint = False)[::-1]]), requires_grad=True)

# Optimization loop
for i in range(steps):
  params = optimizer.step(cost_function, params)
  if i%10 == 0:
    print(f"Step {i}: Cost = {cost_function(params)}")
QAOA_cost = cost_function(params, normalized = False)
print(f"The expectation value of the cost function: {QAOA_cost}, and the performance is: {(100 * QAOA_cost / -432247):.2f} % of the optimal value.")

# Get the final probabilities using optimized parameters
probs = prob_populations(params)

# Plot probability distribution
plt.bar(range(2**wires), probs)
plt.xlabel("Qubit Configuration")
plt.ylabel("Probability")
plt.title("Diabatic QAOA Probability Distribution")
plt.show()

# Find the number of non-zero elements in probs
count = sum([1 for x in probs if x != 0])
print(count)

# Get top (N) indices sorted by highest probability
indices = np.where(probs >= nth_largest(probs, count))[0]
indices = indices[np.argsort(probs[indices])[::-1]]  # Sort by descending probabilities

# Convert to binary representation
binary_indices = [np.binary_repr(i, width=wires) for i in indices]

# Compute costs efficiently using list comprehension
costs = np.array([calculate_cost(b) for b in binary_indices])

# Get the 3 lowest costs and corresponding configurations
threshold = -nth_largest(-costs, 3)
final_mask = costs <= threshold

final_indices = np.array(binary_indices)[final_mask]
final_costs = costs[final_mask]
final_probs = probs[indices[final_mask].astype(int)]  # Get corresponding probabilities

# Print the results
for config, cost, prob in zip(final_indices, final_costs, final_probs):
    print(f"config: {config}, cost: {cost}, prob: {100*prob:.8f}%")

# Sum all the probabilities to yield the correct value (without considering the slack variables)
total_prob = 0
for index, prob in enumerate(probs):
  if check_first_five_bits(index):
    total_prob += prob
print(f"The probability to get the correct value is: {100 *total_prob}%")
```

```python
# To outline the probability for each of the variable configuartion (without the slack variables)

five_bits_probs = defaultdict(int)

for index, prob in enumerate(probs):
  five_bits_probs[index//2**(wires-5)] += prob

plt.plot(five_bits_probs.keys(), five_bits_probs.values())
values_list = list(five_bits_probs.values())
max_index = values_list.index(max(values_list))
max_key = list(five_bits_probs.keys())[max_index]

print(f"The probability to get the correct minimum: {100*five_bits_probs[30]}%")
print(f"The most probable value is at {np.binary_repr(max_key)} with the probability of {100* five_bits_probs[max_key]}%") # This gives the second best result, for now.
```

# Custom Mixer Layer
An original idea & finds the optimum point faster with more parameters. At the  end of the day, gradient optimizers can never escape local minima.

```python
# Custom mixer per layer
def cm_mixer_hamiltonian_layer(a, b):
    coefficients = []
    operators = []
    for w in range(wires):
        coefficients.append(-a)  # X term
        operators.append(qml.PauliX(w))
        coefficients.append(-b)  # Z term
        operators.append(qml.PauliZ(w))
    return qml.Hamiltonian(coefficients, operators)

# QAOA layer with the tunable mixer Hamiltonian
def qaoa_layer_cm(gamma, beta, a, b):
    qaoa.cost_layer(gamma, H_C)
    qaoa.mixer_layer(beta, cm_mixer_hamiltonian_layer(a, b))

def circuit_cm(params, **kwargs):
    for w in range(wires):
        qml.Hadamard(w)
    for d in range(depth):
        qaoa_layer_cm(params[0][d], params[1][d], params[2][d], params[3][d])

@qml.qnode(device)
def cost_function_cm(params, normalized = True):
    circuit_cm(params)
    if normalized == True:
      expval = qml.expval(H_C)  # Ensure `H_C` is differentiable
    else:
      expval = qml.expval(cost_Hamiltonian)
    return expval

@qml.qnode(device)
def prob_populations_cm(params):
  circuit_cm(params)
  return qml.probs(wires = range(wires))
```

```python
# --- Fourier parameterization ---
def build_fourier_params(amps, depth, q):
    """
    Reconstruct gamma, beta, a, b from Fourier amplitudes.
    """
    gammas = np.zeros(depth)
    betas = np.zeros(depth)
    a_params = np.zeros(depth)
    b_params = np.zeros(depth)

    uk = amps[:q]
    vk = amps[q:2*q]
    ak = amps[2*q:3*q]
    bk = amps[3*q:4*q]

    positions = (np.arange(1, depth + 1) - 0.5) / depth
    for k in range(1, q + 1):
        factor = np.pi * (k - 0.5)
        gammas += uk[k - 1] * np.sin(factor * positions)
        betas += vk[k - 1] * np.cos(factor * positions)
        a_params += ak[k - 1] * np.cos(factor * positions)
        b_params += bk[k - 1] * np.sin(factor * positions)

    # # Optional normalization of mixer terms
    # total = a_params + b_params + 1e-8
    # a_params = a_params / total
    # b_params = b_params / total

    return np.vstack([gammas, betas, a_params, b_params])

# --- Cost function wrapper ---
def cost_function_fourier(amps, normalized = True):
    params = build_fourier_params(amps, depth, q)
    return cost_function_cm(params, normalized)

# --- Hyperparameters ---
depth = 10
steps = 200
stepsize = 0.1
q = 3  # Number of Fourier modes
num_restarts = 50

# For tracking the best result
all_costs = []
all_amps = []

for restart in range(num_restarts):
    print(f"\n Restart {restart + 1}/{num_restarts}")

    # Initialize Fourier amplitudes (random)
    amps = np.hstack([
        np.random.uniform(0, np.pi/2, q),   # gamma amplitudes (u_k)
        np.random.uniform(0, np.pi/4, q),   # beta amplitudes (v_k)
        np.random.uniform(0, 1, q),         # a amplitudes (a_k)
        np.random.uniform(0, 1, q)          # b amplitudes (b_k)
    ])
    amps = np.array(amps, requires_grad=True)

    optimizer = qml.AdamOptimizer(stepsize=stepsize)

    for step in range(steps):
        amps = optimizer.step(cost_function_fourier, amps)
        amps = np.clip(amps, 0, np.pi)
        if step % 50 == 0:
            cost = cost_function_fourier(amps)
            print(f"Step {step}: Cost = {cost}")

    final_cost = cost_function_fourier(amps, normalized = False)
    print(f" Local minimum cost: {final_cost}")

    all_costs.append(final_cost)
    all_amps.append(amps)

# Best result
best_idx = np.argmin(all_costs)
best_cost = all_costs[best_idx]
best_amps = all_amps[best_idx]
best_params = build_fourier_params(best_amps, depth, q)

print(f"\n Best cost found over {num_restarts} restarts: {best_cost}")
print(f"Performance: {(100 * best_cost / -432247):.2f}% of the optimal value.")

# --- Final probability distribution ---
probs = prob_populations_cm(best_params)

# Plot probability distribution
plt.bar(range(2**wires), probs)
plt.xlabel("Qubit Configuration")
plt.ylabel("Probability")
plt.title("Diabatic QAOA Probability Distribution (Best Run)")
plt.show()

# --- Top configurations ---
count = sum([1 for x in probs if x != 0])
indices = np.where(probs >= nth_largest(probs, count))[0]
indices = indices[np.argsort(probs[indices])[::-1]]
binary_indices = [np.binary_repr(i, width=wires) for i in indices]
costs = np.array([calculate_cost(b) for b in binary_indices])

threshold = -nth_largest(-costs, 3)
final_mask = costs <= threshold

final_indices = np.array(binary_indices)[final_mask]
final_costs = costs[final_mask]
final_probs = probs[indices[final_mask].astype(int)]

for config, cost, prob in zip(final_indices, final_costs, final_probs):
    print(f"config: {config}, cost: {cost}, prob: {100*prob:.8f}%")

total_prob = 0
for index, prob in enumerate(probs):
    if check_first_five_bits(index):
        total_prob += prob
print(f"The probability to get the correct value is: {100 * total_prob}%")
```

# Full Basin-Hopping QAOA

```python
from scipy.optimize import basinhopping

depth = 7
# Initialize parameters
gammas = np.linspace(np.pi/2, 0, depth, endpoint=False)[::-1]
betas = np.linspace(np.pi/4, 0, depth)
a_params = np.linspace(1, 0, depth)
b_params = np.linspace(0, 1, depth)

# Flatten all parameters into a single array
initial_params = np.vstack([gammas, betas, a_params, b_params]).flatten()

# Objective function for basin-hopping
def objective(flat_params):
    params = flat_params.reshape(4, depth)
    cost = cost_function_cm(params)
    counter[0] += 1
    if counter[0] % 10 == 0:
        print(f"Evaluation {counter[0]}: cost = {cost}")
    return cost

# Callback to track cost
cost_history = []

def callback(x, f, accept):
    print(f"Cost: {f:.6f}, Accepted: {accept}")
    cost_history.append(f)

# Run basin-hopping optimization
minimizer_kwargs = {
    "method": "L-BFGS-B",
    "options": {"maxiter": 20}  # Limit inner optimizer work
}

result = basinhopping(
    objective,
    initial_params,
    niter=10,              # Number of basin-hopping steps
    minimizer_kwargs=minimizer_kwargs,
    stepsize=0.1,           # Step size of random jumps
    callback=callback,
    disp=True
)

# Extract optimized parameters
optimized_params = result.x.reshape(4, depth)

# Final cost evaluation
final_cost = cost_function_cm(optimized_params, normalized=False)
print(f"\nFinal unnormalized cost: {final_cost}")
print(f"Performance: {(100 * final_cost / -432247):.2f}% of the optimal value.")

# Plot cost over iterations
plt.plot(cost_history)
plt.xlabel('Basin-Hopping Step')
plt.ylabel('Cost')
plt.title('Cost over Basin-Hopping Optimization')
plt.show()

# Get final probabilities
probs = prob_populations_cm(optimized_params)

# Plot probability distribution
plt.bar(range(2**wires), probs)
plt.xlabel("Qubit Configuration")
plt.ylabel("Probability")
plt.title("Diabatic QAOA Probability Distribution")
plt.show()

# Find the number of non-zero probabilities
count = sum([1 for x in probs if x != 0])

# Get top (N) indices sorted by highest probability
indices = np.where(probs >= nth_largest(probs, count))[0]
indices = indices[np.argsort(probs[indices])[::-1]]  # Sort by descending probabilities

# Convert indices to binary representations
binary_indices = [np.binary_repr(i, width=wires) for i in indices]

# Compute costs for top configurations
costs = np.array([calculate_cost(b) for b in binary_indices])

# Get the 3 lowest costs and corresponding configurations
threshold = -nth_largest(-costs, 3)
final_mask = costs <= threshold

final_indices = np.array(binary_indices)[final_mask]
final_costs = costs[final_mask]
final_probs = probs[indices[final_mask].astype(int)]

# Print the best configurations
for config, cost, prob in zip(final_indices, final_costs, final_probs):
    print(f"config: {config}, cost: {cost}, prob: {100*prob:.8f}%")

# Compute total probability of correct solutions (based on first 5 bits)
total_prob = 0
for index, prob in enumerate(probs):
    if check_first_five_bits(index):
        total_prob += prob
print(f"The probability to get the correct value is: {100 * total_prob}%")
```

# Spectral Extraction for Optimization Problems

```python
from scipy.fft import fft, fftfreq

# Parameters
wires = len(Q)
device = qml.device('default.qubit', wires=wires)

# Time evolution parameters
num_time_points = 200
t_max = 20
times = np.linspace(0, t_max, num_time_points)

# QAOA cost Hamiltonian
cost_Hamiltonian, offset, coeff_max = create_cost_Hamiltonian(Q=Q, converted=0)

# State preparation and time evolution circuit
@qml.qnode(device)
def evolve_and_measure(t):
    # Prepare the |+>^n state
    for w in range(wires):
        qml.Hadamard(w)
    # Evolve under the cost Hamiltonian for time t
    qml.ApproxTimeEvolution(cost_Hamiltonian, t, 1)
    # Measure expectation value of the cost Hamiltonian
    return qml.expval(cost_Hamiltonian)

# Collect expectation values over time
expectation_values = np.array([evolve_and_measure(t) for t in times])

# Plot time-domain expectation values
plt.plot(times, expectation_values)
plt.xlabel('Time')
plt.ylabel('Cost expectation value')
plt.title('Time evolution under cost Hamiltonian')
plt.show()

# FFT to extract frequencies
freqs = fftfreq(num_time_points, d=(times[1] - times[0]))
fft_values = fft(expectation_values)

# Plot frequency spectrum
plt.plot(freqs[:num_time_points // 2], np.abs(fft_values)[:num_time_points // 2])
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('Frequency spectrum of cost expectation value')
plt.show()
```
